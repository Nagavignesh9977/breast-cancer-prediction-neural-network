```python
import numpy as np
import pandas as pd
import sklearn.datasets
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
dataframe = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
dataframe.head()
```

```python
dataframe['label'] = breast_cancer_dataset.target
```

```python
dataframe.tail()
```

```python
dataframe.describe()
```

```python
X = dataframe.drop(columns='label', axis=1)
Y = dataframe['label']
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state =2,test_size =0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(X_train)
x_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting the input ,hidden,output layers
model = keras.Sequential([
    keras.layers.Flatten(input_shape =(30,)),
    keras.layers.Dense(20,activation = 'relu'),
    keras.layers.Dense(2,activation = 'sigmoid')
])
```

```python
model.compile(optimizer ="adam",
              loss = "sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
history = model.fit(x_train,Y_train,validation_split=0.1,epochs=10)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("model_accuracy")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend(["training_data","validation_data"],loc = "lower right")
```

```python

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("model_loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(["training_data","validation_data"],loc = "lower right")
```

```python
 loss,accuracy = model.evaluate(x_test,Y_test)
 print(accuracy)
```
